# pre processing

```python
df=pd.get_dummies(book)
df.head()
```

# apriori algorithm

```python
ferquent_itemsets= apriori(df,min_support=0.1,use_colnames=True)
ferquent_itemsets
```

```python
rules = association_rules(ferquent_itemsets,metric="lift", min_threshold=0.7)
rules
rules.sort_values('lift',ascending=False)[0:20]
```

```python
rules[rules.lift>1]
```
